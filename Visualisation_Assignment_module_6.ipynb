```python
# Question Number 5 >>
# Show a pie chart to represent the percentage distribution of different sections in the array `sections`.

# sections = ['Section A', 'Section B', 'Section C', 'Section D']
# sizes = [25, 30, 15, 30]

# Answer >>>

sections = ['Section A', 'Section B', 'Section C', 'Section D']
sizes = [25, 30, 15, 30]

plt.pie(sizes, labels=sections, autopct='%1.1f%%', startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures >>
plt.title('Percentage Distribution of Sections')
plt.show()
```

```python
# Question Number 5 >>>
# Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

# Answer >>>

# Generate a synthetic dataset with correlated features
np.random.seed(42)
data = np.random.rand(100, 4)
df = pd.DataFrame(data, columns=['Feature 1', 'Feature 2', 'Feature 3', 'Feature 4'])

#  some correlation b/w features
df['Feature 2'] = df['Feature 1'] + np.random.randn(100) * 0.5
df['Feature 3'] = df['Feature 1'] + df['Feature 2'] + np.random.randn(100) * 0.2


# Calculate the correlation matrix >>
correlation_matrix = df.corr()

# Visualize the correlation matrix using a heatmap >>
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# Question Number 4 >>>
#  Using the given x and y data, generate a 3D surface plot to visualize the function z = sin((x**2 + y**2)**(1/2))

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Answer >>>

import numpy as np
import pandas as pd
import plotly.graph_objects as go

# Create 3D surface plot
fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])
fig.update_layout(title='3D Surface Plot', scene = dict(
                    xaxis_title='X',
                    yaxis_title='Y',
                    zaxis_title='Z'),
                    autosize=False,
                    width=700, height=700)

fig.show()
```

```python
# Question Number 1 >>>
# Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

# Answer >>>

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

output_notebook()

# Set x-values from 0 to 10
x = np.linspace(0, 10, 100)
# y-values as the sine of x
y = np.sin(x)

# Create a new plot >>
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='y')
p.line(x, y, legend_label="y=sin(x)", line_width=2)

show(p)
```

```python
# Question Nymber 2 >>>
# Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

# Answer >>>

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

output_notebook()

# Randomly generated x and y values >.>
x = np.random.rand(100)
y = np.random.rand(100)
sizes = np.random.randint(10, 80, 100)
colors = np.random.choice(['red', 'green', 'blue', 'purple', 'orange'], 100)

# Create a scatter plot >>>
p = figure(title="Scatter Plot", x_axis_label='x', y_axis_label='y')

# Add circle renderers >>
p.scatter(x, y, size=sizes, color=colors, alpha=0.5)

show(p)
```

```python
# Question Number 3 >>>
# Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Answer >>>

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import factor_cmap

output_notebook()

# Create a bar chart
p = figure(x_range=fruits, title="Fruit Counts", toolbar_location=None, tools="")

p.vbar(x=fruits, top=counts, width=0.9, color=factor_cmap('x', palette=['#c9d9d3', '#718dbf', '#e84d60', '#ddb7b1'], factors=fruits))

show(p)
```

```python
# Question Number 4 >>>
# Create a Bokeh histogram to visualize the distribution of the given data.
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Answer >>>

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

output_notebook()

# Create a histogram >>>
p = figure(title="Histogram", x_axis_label='Value', y_axis_label='Frequency')

p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="navy", line_color="white", alpha=0.7)

show(p)
```

```python
# Question Number 5 >>>
#  Create a Bokeh heatmap using the provided dataset.
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Answewr >>>

from bokeh.plotting import figure, show
from bokeh.models import LinearColorMapper, ColorBar, BasicTicker
from bokeh.transform import transform
from bokeh.io import output_notebook
import numpy as np

output_notebook()

# Create a heatmap
p = figure(title="Heatmap", x_axis_label='X-axis', y_axis_label='Y-axis', tooltips=[("Value", "@image")])

color_mapper = LinearColorMapper(palette='Viridis256', low=data_heatmap.min(), high=data_heatmap.max())

p.image(image=[data_heatmap], x=0, y=0, dw=1, dh=1, color_mapper=color_mapper)

color_bar = ColorBar(color_mapper=color_mapper, ticker=BasicTicker(), location=(0, 0))
p.add_layout(color_bar, 'right')

show(p)
```
